### Gráficos

```python
algoritmos = df_algoritmos['Algoritmo'].unique()
```

```python
for algoritmo in algoritmos:
    # Filtrar el DataFrame por algoritmo
    df_filtrado = df_algoritmos[df_algoritmos['Algoritmo'] == algoritmo]

    # Crear el gráfico de barras
    plt.figure(figsize=(8, 4))
    plt.bar(df_filtrado['Tipo de Medición'], df_filtrado['Promedio General'], color='skyblue')

    # Añadir título y etiquetas a los ejes
    plt.title(f'Promedio General por Tipo de Medición - {algoritmo}')
    plt.xlabel('Algoritmo de Medición')
    plt.ylabel('Promedio General (segundos)')

    # Mostrar el gráfico
    plt.show()
```

#### Gráficos de palabras

```python
palabras = df_palabras['Palabra'].unique()
```

```python
for palabra in palabras:
    df_palabra = df_palabras[df_palabras['Palabra'] == palabra]
    plt.figure(figsize=(14, 8))
    sns.barplot(data=df_palabra, x='Algoritmo', y='Promedio Tiempo', hue='Tipo de Medición', ci=None)
    plt.title(f'Comparación de tiempos de ejecución para la palabra: {palabra}')
    plt.xlabel('Algoritmo de Medición')
    plt.ylabel('Promedio Tiempo')
    plt.legend(title='Tipo de Medición')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Comparación algoritmos

```python
df_timeit = df_algoritmos[df_algoritmos['Tipo de Medición'] == 'timeit']
df_perf_counter = df_algoritmos[df_algoritmos['Tipo de Medición'] == 'perf_counter']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_timeit, x='Algoritmo', y='Promedio General', hue='Algoritmo', dodge=False, palette='viridis')
plt.title('Tiempos de Ejecución Promedio por Algoritmo (timeit)')
plt.xlabel('Algoritmo')
plt.ylabel('Promedio General')
plt.legend().set_visible(False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_perf_counter, x='Algoritmo', y='Promedio General', hue='Algoritmo', dodge=False, palette='viridis')
plt.title('Tiempos de Ejecución Promedio por Algoritmo (perf_counter)')
plt.xlabel('Algoritmo')
plt.ylabel('Promedio General')
plt.legend().set_visible(False)
plt.show()
```

### Desviación y Distribución

```python
mean = df_algoritmos['Promedio General'].mean()
std = df_algoritmos['Promedio General'].std()
```

```python
x = np.linspace(mean - 3*std, mean + 3*std, 100)
y = norm.pdf(x, mean, std)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', lw=2, label='Distribución Normal')
plt.title('Distribución Normal de los Tiempos de Ejecución Promedio')
plt.xlabel('Promedio General')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
unique_combinations = df_palabras[['Algoritmo', 'Tipo de Medición']].drop_duplicates()
```

```python
for index, row in unique_combinations.iterrows():
    alg = row['Algoritmo']
    tipo_medicion = row['Tipo de Medición']
    
    # Filtrar el DataFrame para la combinación actual
    df_filtered = df_palabras[(df_palabras['Algoritmo'] == alg) & (df_palabras['Tipo de Medición'] == tipo_medicion)]
    
    # Calcular la media y la desviación estándar
    mean = df_filtered['Promedio Tiempo'].mean()
    std = df_filtered['Promedio Tiempo'].std()
    
    # Generar los datos para la distribución normal
    x = np.linspace(mean - 3*std, mean + 3*std, 100)
    y = norm.pdf(x, mean, std)
    
    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'r-', lw=2, label='Distribución Normal')
    plt.title(f'Distribución Normal de los Tiempos de Ejecución Promedio para {alg} con {tipo_medicion}')
    plt.xlabel('Promedio General')
    plt.ylabel('Densidad')
    plt.legend()
    plt.show()
```
