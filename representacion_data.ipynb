### Desviaci贸n y Distribuci贸n

```python
mean = df_algoritmos['Promedio General'].mean()
std = df_algoritmos['Promedio General'].std()
```

```python
x = np.linspace(mean - 3*std, mean + 3*std, 100)
y = norm.pdf(x, mean, std)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', lw=2, label='Distribuci贸n Normal')
plt.title('Distribuci贸n Normal de los Tiempos de Ejecuci贸n Promedio')
plt.xlabel('Promedio General')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
unique_combinations = df_palabras[['Algoritmo', 'Tipo de Medici贸n']].drop_duplicates()
```

```python
for index, row in unique_combinations.iterrows():
    alg = row['Algoritmo']
    tipo_medicion = row['Tipo de Medici贸n']
    
    # Filtrar el DataFrame para la combinaci贸n actual
    df_filtered = df_palabras[(df_palabras['Algoritmo'] == alg) & (df_palabras['Tipo de Medici贸n'] == tipo_medicion)]
    
    # Calcular la media y la desviaci贸n est谩ndar
    mean = df_filtered['Promedio Tiempo'].mean()
    std = df_filtered['Promedio Tiempo'].std()
    
    # Generar los datos para la distribuci贸n normal
    x = np.linspace(mean - 3*std, mean + 3*std, 100)
    y = norm.pdf(x, mean, std)
    
    # Crear el gr谩fico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'r-', lw=2, label='Distribuci贸n Normal')
    plt.title(f'Distribuci贸n Normal de los Tiempos de Ejecuci贸n Promedio para {alg} con {tipo_medicion}')
    plt.xlabel('Promedio General')
    plt.ylabel('Densidad')
    plt.legend()
    plt.show()
```
